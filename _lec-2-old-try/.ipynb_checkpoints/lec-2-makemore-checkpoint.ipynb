```python
words = open('names.txt' , 'r').read().splitlines()
```

```python
words
```

```python
bigrams = {}
for word in words:
    chs = ['<S>'] + list(word) + ['<E>']
    print(chs)
    for ch1,ch2 in zip(chs ,chs[1:]):
        bigram = (ch1,ch2)
        bigrams[bigram] = bigrams.get(bigram,0) + 1
```

```python
bigrams
```
